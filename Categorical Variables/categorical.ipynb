### Using Mapping

```python
mapping_ord2 = {
    'Cold' : 0,
    'Hot' : 1, 
    'Lava Hot' : 2, 
    'Boiling Hot' : 3,
    'Freezing' : 4, 
    'Warm' : 5
}
df.ord_2.map(mapping_ord2)
```

### Using LableEncoder

```python
from sklearn import preprocessing

df.loc[:, 'ord_2'] = df.ord_2.fillna('NONE')

lbl_encoder = preprocessing.LabelEncoder()

df.loc[:, 'ord_2'] = lbl_encoder.fit_transform(df.ord_2.values)
```

```python
df
```

### Using BinaryEncoder

```python
from scipy import sparse
import numpy as np

binarizer = preprocessing.LabelBinarizer()

df.loc[:, 'ord_1'] = lbl_encoder.fit_transform(df.ord_1.values)
binarizer.fit_transform(df.ord_1.values).tolist()
```

### Using One-Hot Encoder

```python

from sklearn import preprocessing

example = np.random.randint(1000, size = 1000000)
```

```python
ohencoder = preprocessing.OneHotEncoder(sparse=False)

ohe_example = np.array(example.reshape(-1, 1))
unsparse_output = ohencoder.fit_transform(ohe_example)
```

```python
unsparse_output.shape
```

```python
unsparse_output
```

```python
ohencoder = preprocessing.OneHotEncoder(sparse=True)
ohe_example = np.array(example.reshape(-1, 1))
sparse_output = ohencoder.fit_transform(ohe_example)
sparse_output
```
